Pseudocódigo:

Inicio

    Para cada tolerancia t en {1e-3, 1e-10}

        x ← 1

            termino ← ((-1)**(n+1)) * (x**(2*n-1)) / (2*n - 1)

```python
import math

for tolerancia in [1e-3, 1e-10]:
    suma = 3
    n = 2
    signo = 1
    error = 1
    while error > tolerancia:
        termino = signo * (4 / (n*(n+1)*(n+2)))
        suma += termino
        error = abs(suma - math.pi)
        signo *= -1
        n += 2
    print(f"Tolerancia {tolerancia}")
    print("Aproximación de pi:", suma)
    print("Error:", error)
    print("Términos necesarios:", (n-2)//2)
    print()
```

Pseudocódigo:

Algoritmo MachinPi

```python
import math

n = 1
suma = 0
error = 1

while error > 1e-3:
    termino = ((-1)**(n+1)) * (4*(1/5)**(2*n-1)/(2*n-1) - (1/239)**(2*n-1)/(2*n-1))
    suma += termino
    error = abs(4*suma - math.pi)
    n += 1

print("Términos necesarios para error < 10⁻³:", n)
print("Aproximación de π:", 4*suma)
print("Error:", error)
```

1. Escriba un algoritmo para sumar la serie finita $\sum_{i=1}^n x_i$ en orden inverso.

ALGORITMO suma_inversa:

    ENTRADA: N (tamaño), X (una lista de números x_1, ..., x_N)
    SALIDA: SUM
    
    Paso 1: SUM = 0.0
    Paso 2: PARA i = N HASTA 1 (con paso -1):
    Paso 3:     SUM = SUM + X[i]  // (Asumiendo 1-indexado)
    Paso 4: SALIDA (SUM)
    Paso 5: PARE.

```python
def suma_inversa(lista_x):
    SUM = 0.0
    for x_i in reversed(lista_x):
        SUM += x_i
    return SUM


x = [1, 2, 3, 4, 5]
print(f"Suma inversa: {suma_inversa(x)}")
```

La suma inversa reduce el error numérico.

Escriba y ejecute un algoritmo que determine el número de términos necesarios en el lado izquierdo de la ecuación de tal forma que el lado izquierdo difiera del lado derecho en menos de $10^{-6}$.

Inicio

    x ← 0.25

    derecha ← (1 + 2*x) / (1 + x + x^2)

    izquierda ← 0

    termino ← (2^i * x^(2*i)) / (1 - x^i + x^(2*i))
