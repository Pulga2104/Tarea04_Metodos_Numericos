    b. $\Sigma_{i=1}^{10}(\frac{1}{i^{3}})$ primero por $\frac{1}{1}+\frac{1}{8}+\frac{1}{27}+\cdot\cdot\cdot+\frac{1}{1000}$ y luego por $\frac{1}{1000}+\frac{1}{729}+\cdot\cdot\cdot+\frac{1}{1}$ 5

    Pseudocódigo:

    Inicio

            suma1 ← 0

            Para i ← 1 Hasta 10 Hacer

                suma1 ← suma1 + truncar(1 / (i*i), 3)

```python
def truncar(num, digitos):
    return float(str(num)[:str(num).find('.') + digitos + 1])

suma1_a = 0
suma2_a = 0
for i in range(1, 11):
    suma1_a += truncar(1 / (i**2), 3)
for i in range(10, 0, -1):
    suma2_a += truncar(1 / (i**2), 3)

print("Literal a")
print("Suma directa:", suma1_a)
print("Suma inversa:", suma2_a)

suma1_b = 0
suma2_b = 0
for i in range(1, 11):
    suma1_b += truncar(1 / (i**3), 3)
for i in range(10, 0, -1):
    suma2_b += truncar(1 / (i**3), 3)

print("\nLiteral b")
print("Suma directa:", suma1_b)
print("Suma inversa:", suma2_b)
```

(a) La suma inversa (1.550) es más precisa porque reduce el error al sumar primero los valores pequeños.

(b) También en este caso la suma inversa (1.198) es más precisa por la misma razón.

$$arctan~x=\lim_{n\rightarrow\infty}P_{n}(x)=\lim_{n\rightarrow\infty}\sum_{l=1}^{n}(-1)^{l+1}\frac{x^{2l-1}}{2i-1}$$

b. El lenguaje de programación $C++$ requiere que el valor de $\pi$ se encuentre dentro de $10^{-10}.$ ¿Cuántos términos de la serie se necesitarían sumar para obtener este grado de precisión?

Pseudocódigo:

Inicio

    Para cada tolerancia t en {1e-3, 1e-10}

        x ← 1

        n ← 1

        suma ← 0

        repetir

            termino ← ((-1)**(n+1)) * (x**(2*n-1)) / (2*n - 1)

            suma ← suma + termino

            error ← abs(4*suma - 3.1416)

            n ← n + 1

        hasta que error < t

        Imprimir "Términos necesarios para tolerancia t:", n

    FinPara
    
Fin


```python
import math

for tolerancia in [1e-3, 1e-10]:
    suma = 3
    n = 2
    signo = 1
    error = 1
    while error > tolerancia:
        termino = signo * (4 / (n*(n+1)*(n+2)))
        suma += termino
        error = abs(suma - math.pi)
        signo *= -1
        n += 2
    print(f"Tolerancia {tolerancia}")
    print("Aproximación de pi:", suma)
    print("Error:", error)
    print("Términos necesarios:", (n-2)//2)
    print()
```

(a) Se necesitan 6 términos para error < 10⁻³.

(b) Se necesitan aproximadamente 1357 términos para error < 10⁻¹⁰.

3. Otra fórmula para calcular $\pi$ se puede deducir a partir de la identidad $\pi/4=4 \text{ arctan}(1/5) - \text{arctan}(1/239)$. Determine el número de términos que se deben sumar para garantizar una aproximación dentro de $10^{-3}$

Pseudocódigo:

Algoritmo MachinPi

Inicio

    n ← 1

    suma ← 0

    repetir

        termino ← ((-1)**(n+1)) * (4*(1/5)**(2*n-1)/(2*n-1) - (1/239)**(2*n-1)/(2*n-1))

        suma ← suma + termino

        error ← abs(4*suma - 3.1416)

        n ← n + 1

    hasta que error < 0.001

    Imprimir "Términos:", n
    
Fin


```python
import math

n = 1
suma = 0
error = 1

while error > 1e-3:
    termino = ((-1)**(n+1)) * (4*(1/5)**(2*n-1)/(2*n-1) - (1/239)**(2*n-1)/(2*n-1))
    suma += termino
    error = abs(4*suma - math.pi)
    n += 1

print("Términos necesarios para error < 10⁻³:", n)
print("Aproximación de π:", 4*suma)
print("Error:", error)
```

Se necesitan 3 términos para un error menor que 10⁻³.
